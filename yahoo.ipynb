```python
import requests
from bs4 import BeautifulSoup

def scrape_yahoo_finance_community(stock_symbol):
    url = f'https://finance.yahoo.com/quote/{stock_symbol}/community/'
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        comments = []
        
        # 'Sort by' 드롭다운 클릭
        sort_by_dropdown = driver.find_element(By.XPATH, '//*[@id="sort-by-select"]/button')
        sort_by_dropdown.click()
        time.sleep(1)

        # 'Newest' 선택
        newest_option = driver.find_element(By.XPATH, '//*[@id="sort-by-select"]/div/ul/li[2]/button')
        newest_option.click()
        time.sleep(5)  # 댓글 로딩 대기

        # Find all comments on the page
        comment_blocks = driver.find_elements(By.XPATH, '//li[contains(@class, "C($c-fuji-grey-j) Mb(4px) Lh(1.7)")]')

        for block in comment_blocks:
            comment = block.get_text(strip=True)
            comments.append(comment)

        return comments
    
    else:
        print(f"Failed to fetch data: {response.status_code}")
        return None


# Example usage
stock_symbol = 'NVDA'
comments = scrape_yahoo_finance_community(stock_symbol)

if comments:
    for i, comment in enumerate(comments, start=1):
        print(f"Comment {i}: {comment}\n")
else:
    print("No comments found or failed to scrape the data.")
```
